**2.Which sunscreen ingredients have the highest penetration rates into the bloodstream, and what are the health effects?**

**Description**:Sunscreen Ingredients like Oxybenzone, Octinoxate, Homosalate, Octocrylene, PABA, 4-MBC, Avobenzone, Benzophenone, Benzyl Salicylate are known to have high penetration rates into the bloodstream. These chemicals can pass through the skin and enter the bloodstream, leading to potential systemic exposure.
These ingredients  have been linked to potential health risks like hormone disruption, endocrine disruption, cell damage, allergic reactions, carcinogenic potential and skin irritation. Studies also suggest they may affect the endocrine system.
Oxybenzone (Benzophenone-3):Oxybenzone has been identified as a potential endocrine disruptor because it can be absorbed through the skin and may act like an estrogen (a hormone) in the body. This means it could theoretically interfere with the normal hormonal balance, particularly estrogen levels.Oxybenzone concentrations in sunscreen range from 2% to 6%.
Octinoxate (Ethylhexyl Methoxycinnamate)::octinoxate can mimic the hormone estrogen in the body, which can lead to endocrine disruption at high concentrations.Octinoxate is commonly used in sunscreens at concentrations of 7.5% to 10%.
PABA (Para-aminobenzoic acid)::Itchy, red skin at the sunscreen application site.PABA was typically used in concentrations of 5% to 15% in sunscreen formulations in the past.5% to 10% was a common concentration range for PABA-containing sunscreens.
Cinnamate Derivatives (e.g., Octocrylene):cinnamate derivatives have been linked to skin irritation and redness in some individuals.Octocrylene is often used in sunscreens in concentrations typically ranging from 1% to 10%.Typical concentration: 2% to 5% is commonly found in many sunscreen products.
Retinoids and Alpha Hydroxy Acids (AHAs):AHAs exfoliate the skin, which removes dead skin cells and makes the new skin underneath more sensitive to UV rays.Retinoids accelerate skin cell turnover, which similarly leads to thinner, more delicate skin.
When the usage of sunscreen with excessive amounts of AHAs or retinoids, the skin becomes more prone to sunburn, hyperpigmentation, and premature aging because it loses some of its natural ability to protect itself from UV radiation.AHAs (usually around 2% to 5%),retinoids (0.1-1%).

```python
data["Ingredients"] = data["Ingredients"].astype(str)
harmful_ingredients = ["Oxybenzone", "Octinoxate", "Homosalate", "Octocrylene", 
                       "Avobenzone", "PABA", "4-MBC", "Benzophenone", "Benzyl Salicylate"]
harmful_counts = {ingredient: data["Ingredients"].str.contains(ingredient, case=False, na=False).sum()
                  for ingredient in harmful_ingredients}

total_products = len(data) 

harmful_percentages = {ingredient: (count / total_products) * 100 for ingredient, count in harmful_counts.items()}


print("Harmful Ingredients Percentage:", harmful_percentages)


plt.figure(figsize=(8, 5))
sns.barplot(x=list(harmful_percentages.keys()), y=list(harmful_percentages.values()), palette="coolwarm")


plt.xticks(rotation=45)
plt.ylabel("Percentage (%)")
plt.xlabel("Harmful Ingredients")
plt.title("Percentage of Products Containing Harmful Ingredients")


for index, value in enumerate(harmful_percentages.values()):
    plt.text(index, value + 0.5, f"{value:.1f}%", ha="center", fontsize=10)

plt.show()
```

```python
harmful_ingredients = {
    "Oxybenzone": "Hormone Disruption",
    "Octinoxate": "Endocrine Disruption",
    "Homosalate": "Hormone Disruption",
    "Octocrylene": "Cell Damage",
    "Avobenzone": "Skin Irritation",
    "PABA": "Allergic Reactions",
    "4-MBC": "Hormone Disruption",
    "Benzophenone": "Carcinogenic Potential",
    "Benzyl Salicylate": "Skin Allergies"
}
harmful_counts = {ingredient: data["Ingredients"].str.contains(ingredient, case=False, na=False).sum()
                  for ingredient in harmful_ingredients.keys()}

disease_contribution = {}
for ingredient, effect in harmful_ingredients.items():
    if harmful_counts[ingredient] > 0:
        disease_contribution[effect] = disease_contribution.get(effect, 0) + harmful_counts[ingredient]

total_occurrences = sum(disease_contribution.values())
disease_contribution = {effect: (count / total_occurrences) * 100 for effect, count in disease_contribution.items()}


plt.figure(figsize=(8, 5))
sns.barplot(x=list(disease_contribution.keys()), y=list(disease_contribution.values()), palette="coolwarm")
plt.xticks(rotation=45)
plt.ylabel("Estimated Disease Contribution (%)")
plt.xlabel("Health Effects")
plt.title("Estimated Disease Risk Contribution by Sunscreen Ingredients")
for index, value in enumerate(disease_contribution.values()):
    plt.text(index, value + 0.5, f"{value:.2f}%", ha="center", fontsize=10)
plt.show()
```

**Conclusion**:
The graph "Percentage of Products Containing Harmful Ingredients" shows the information about the percentage of the products which contains the harmful ingredients like "Avobenzone" as the highest and "4-MBC" as the lowest. The graph "Estimated Disease Risk Contribution by Sunscreen Ingredients" shows the information about the percentage of the health effects.Here, "Hormone Disruption" has the highest percentage of 28.38 compare to other health effects and "Carcinogenic Potential" has the lowest percentage of 1.35 compare to other health effects.

**3.How do nanoparticle-based sunscreens compare in effectiveness and safety to
traditional sunscreens?**

```python
nanoparticle_ingredients = ["Zinc Oxide", "Titanium Dioxide"]
chemical_ingredients = ["Oxybenzone", "Octinoxate", "Homosalate", "Octocrylene", 
                        "Avobenzone", "PABA", "4-MBC", "Benzophenone", "Benzyl Salicylate"]

data["Contains_Nanoparticles"] = data["Ingredients"].apply(lambda x: 
                                    any(nano in x for nano in nanoparticle_ingredients))

data["Contains_Chemicals"] = data["Ingredients"].apply(lambda x: 
                                    any(chem in x for chem in chemical_ingredients))

nano_count = data["Contains_Nanoparticles"].sum()
chemical_count = data["Contains_Chemicals"].sum()

avg_spf_nano = data[data["Contains_Nanoparticles"]]["Market Value ($M)"].mean()
avg_spf_chem = data[data["Contains_Chemicals"]]["Market Value ($M)"].mean()

plt.figure(figsize=(7, 7))
plt.pie([nano_count, chemical_count], labels=["Nanoparticle-based", "Chemical-based"], autopct='%1.1f%%',
        colors=sns.color_palette("coolwarm", 2), startangle=140)
plt.title("Proportion of Sunscreen Types")
plt.show()
```

**Conclusion**:
The above pie chart provides with the information of 55.4% of nanoparticle-based skin products and 44.6% of chemical-based skin products.

**5.Which SPF levels are most commonly linked to allergic reactions or contact
dermatitis?**

**Description**: SPF (Sun Protection Factor) is a measure of how well a sunscreen protects the skin from UVB rays, which cause sunburn and contribute to skin cancer.At 25% exposure, individuals may experience skin redness and mild irritation. With 50% exposure, the risk increases to sunburn and early skin aging. At 75% exposure, prolonged damage can lead to hyperpigmentation and DNA damage. Finally, at 100% exposure, there is a significant risk of skin cancer and potential hormonal disruption.

```python
spf_disease_risk = {
    "25%": "Skin Redness & Mild Irritation",
    "50%": "Sunburn & Early Skin Aging",
    "75%": "Hyperpigmentation & DNA Damage",
    "100%": "Skin Cancer Risk & Hormonal Disruption"
}
data1["SPF_Percentage"] = pd.cut(data1["SPF Level"], bins=[0, 15, 30, 50, 100], labels=["25%", "50%", "75%", "100%"])
spf_counts = data1["SPF_Percentage"].value_counts().sort_index()

plt.figure(figsize=(8, 8))
colors = sns.color_palette("Reds", len(spf_counts))
plt.pie(spf_counts, labels=[f"{spf} ({spf_disease_risk[spf]})" for spf in spf_counts.index], 
        autopct='%1.1f%%', colors=colors, startangle=140)
plt.title("SPF Levels and Associated Health Risks")
plt.show()
```

**Conclusion**:
Based on the SPF factor, the pie chart shows the highest percentage at 52.2% for SPF 100%, which is associated with skin cancer and hormonal disruption. The lowest percentage is 14.0% for SPF 25%, linked to skin redness and mild irritation.
Similarly, SPF 50% accounts for 16.6%, associated with sunburn and early skin aging. SPF 75% is distributed between 16.6% and 17.2%, corresponding to hyperpigmentation and DNA damage.

# Consumer Preferences & Trends

**1.How do sunscreen sales vary across different countries, and which country has the highest number of sunscreen products sold?**

```python
merged_df = pd.merge(data1, data, on="SUNSCREEN_ID", how="inner")
spfs_by_country = merged_df.pivot_table(index="Country", columns="SPF Level", values="Sales (Units)", aggfunc="sum", fill_value=0)
plt.figure(figsize=(12, 6))
sns.heatmap(spfs_by_country, cmap="coolwarm", annot=True, fmt=".0f")
plt.title("SPF Level Preferences Across Different Countries")
plt.xlabel("SPF Level")
plt.ylabel("Country")
plt.show()
```

**Conclusion**:
The heatmap shows that SPF 30 and 50 are the most preferred sunscreen levels across multiple countries. Countries with higher sun exposure, like South Africa and Egypt, exhibit a strong preference for SPF 100. SPF 15 is less popular in many regions, likely due to increased awareness of stronger sun protection. Some countries show extreme outliers in SPF preferences, possibly influenced by climate, consumer awareness, or market regulations.

```python
merged_df = pd.merge(data1, data, on="SUNSCREEN_ID", how="inner")
country_sales = merged_df.groupby("Country")["Sales (Units)"].sum().reset_index()
plt.figure(figsize=(12, 6))
ax = sns.barplot(x="Country", y="Sales (Units)", data=country_sales, palette="viridis")
plt.title("Total Sunscreen Sales by Country")
plt.ylabel("Total Sales (Units)")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
for index, row in country_sales.iterrows():
    ax.text(index, row["Sales (Units)"] / 2, row["Country"], ha="center", va="center", fontsize=10, color="white", fontweight="bold", rotation=90)
plt.show()
```

**Conclusion**:The graph shows that Egypt, Germany, and South Africa have the highest sunscreen sales, indicating strong demand and awareness. Countries with hotter climates, like Pakistan and Singapore, also show significant sales. In contrast, the USA and Ukraine have lower sales, possibly due to market preferences or climate factors. Overall, sunscreen demand appears to be influenced by sun exposure and regional awareness.

Based on the data analysis, it was observed that mineral sunscreens, which primarily contain zinc oxide and titanium dioxide, tend to be gentler on the skin and are less likely to cause irritation. In contrast, chemical sunscreens, containing ingredients like oxybenzone and avobenzone, have been associated with a higher risk of skin irritation and potential hormonal effects in some cases.
